```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# Loading data
df = pd.read_csv('./accel_data_10hz/accel_data_crash_heavy.csv')
sample_rate = 10 # 10 samples per second
total_samples = df.shape[0]

# Create a time array in seconds
df['time'] = np.arange(0, total_samples/sample_rate, 1/sample_rate)
# Calculating the magnitude of acceleration
df['magnitude'] = np.sqrt(df['AccX']**2 + df['AccY']**2)
df['JerkX'] = np.gradient(df['AccX'], df['time'])
df['JerkY'] = np.gradient(df['AccY'], df['time'])
df['JerkZ'] = np.gradient(df['AccZ'], df['time'])
df['JerkMag'] = np.sqrt(df['JerkX']**2 + df['JerkY']**2 + df['JerkZ']**2)
fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(df['time'], df['magnitude'])
ax[0].set_ylabel('Acc Mag')

ax[1].plot(df['time'], df['JerkMag'])
ax[1].set_xlabel('Time [s]')

```
